```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
#Tozeret = "" # @param [""]
CarsNameSelection = widgets.Dropdown(
    options=data_df['tozeret_nm'].unique().tolist(),
    description='Tozeret:'
)
#Kinuy_Mishari = "" # @param [""]
CarsNickSelection = widgets.Dropdown(
    options=[],
    description='Kinuy Mishari:',
    style={'description_width': '115px'},  # Allocate more space for the label
    #layout=Layout(margin='10px 0 0 0')  # Add spacing between dropdowns
)

def update_dropdown_options(change=None):
    selected_name = CarsNameSelection.value
    car_nicknames = data_df[data_df['tozeret_nm'] == selected_name]['kinuy_mishari'].unique() #              
    CarsNickSelection.options = car_nicknames.tolist()

output_area = Output() #  '  

def show_details(change=None):
    selected_name = CarsNameSelection.value
    selected_nickname = CarsNickSelection.value
    selected_row = data_df[(data_df['tozeret_nm'] == selected_name) & (data_df['kinuy_mishari'] == selected_nickname)]

    total_records = selected_row.shape[0]
    unique_ramat_gimur = selected_row['ramat_gimur'].unique()

    unique_ramat_gimur_string = ', '.join(unique_ramat_gimur)

    # Update the output
    with output_area:
        output_area.clear_output()
        print(f"Total Cars: {total_records}")
        print(f"Unique Ramat Gimur: {unique_ramat_gimur_string}")

CarsNameSelection.observe(update_dropdown_options, names='value') #        (   )
CarsNickSelection.observe(show_details, names='value')
display(CarsNameSelection)
display(CarsNickSelection)
show_details()
display(output_area)


import requests
import pandas as pd
pd.set_option("display.max_columns", None)
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)


# Create a DataFrame from the 'records' in the JSON response
data_df = pd.DataFrame(data['result']['records'])

# Set Tozeret with the 'tozeret_nm' column values
print("\n")
print(data_df['tozeret_nm'].tolist())
print(data_df['kinuy_mishari'].tolist())
```

```python
# Tab 1: Data Overview
tab1_content = widgets.Output()
with tab1_content:
    print("Data Overview:")
    display(data_df.describe())  # Summary statistics

# Tab 2: Raw Data
tab2_content = widgets.Output()
with tab2_content:
    print("Raw Data:")
    display(data_df)  # Full DataFrame
# Tab 3: Charts
# Tab 3: Charts
tab3_content = widgets.Output()
with tab3_content:
    print("Charts:")
    plt.clf()  # Clear the current figure
    plt.figure(figsize=(10, 8))  # Increased figure size to accommodate labels
    side_length = 10
    data = 5 + np.random.randn(side_length, side_length)
    data += np.arange(side_length)
    data += np.reshape(np.arange(side_length), (side_length, 1))

    # Create heatmap with labels
    heatmap = sns.heatmap(data,
                         cmap='viridis',
                         annot=True,
                         fmt='.1f',
                         cbar_kws={'label': 'Value Intensity'})
    plt.title('Enhanced Heatmap with Labels')
    plt.xlabel('X Axis Position')
    plt.ylabel('Y Axis Position')
    plt.show()  # Add this to display the plot
    plt.close()  # Close the figure to free memory

    # Create Number of Unique Cars Produced Each Year graph
    #plt.clf()
    plt.figure(figsize=(10, 6))
    cars_per_year = data_df.groupby('shnat_yitzur')['tozeret_nm'].nunique() #              
    cars_per_year.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Number of Unique Cars Produced Each Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Unique Cars')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content, tab3_content])
tabs.set_title(0, 'Tab 1: Data Overview')
tabs.set_title(1, 'Tab 2: Raw Data')
tabs.set_title(2, 'Tab 3: Charts')

# Display Tabs
display(tabs)
```
