#Imports

```python
from google.colab import drive
drive.mount('/content/drive')

import ipywidgets as widgets
```

#File contents into list logic

```python
filepath = '/content/drive/MyDrive/Cloud computing - Daniel Ayash/Tirgul 01/Students.txt'

with open(filepath, "r") as file:
  contents = file.read()
file.close()

lines = contents.splitlines()

Students = []

for line in lines:
  student_data = line.split(", ")
  Students.append(student_data)

print(Students)
```

#Making of dropdown and content boxes

```python
# @title Students form

StudentNameSelection = widgets.Dropdown(
    options = [student[0] for student in Students],
    description='Student:',
    disabled=False,
)

email_text = widgets.Text(description='Email:', disabled=True)
courses_text = widgets.Text(description='Courses:', disabled=True)
link_text = widgets.Text(description='Link:', disabled=True)

favorite_tvshow_text = widgets.Text(description='TV Show:', disabled=False)
add_button = widgets.Button(description="Add TV Show")


def StudentInfoChange(change=None):
  student_name = StudentNameSelection.value
  for student in Students:
    if student[0] == student_name:
      email_text.value = student[1]
      fixed_courses = student[2].replace(" ", ",")
      courses_text.value = fixed_courses
      link_text.value = student[3]

def addFavoriteTVshow(button):
  student_name = StudentNameSelection.value
  favorite_tv = favorite_tvshow_text.value

  for student in Students:
    if student[0] == student_name:
      if favorite_tv == "":
        if len(student) > 4:
          student.pop(4)

      else:
        if len(student) > 4:
          student[4] = "Favorite TV show: " + favorite_tv
        else:
          student.append("Favorite TV show: " + favorite_tv)

  with open(filepath, "w") as file:
    for student in Students:
      updated_student_data = ", ".join(student)
      file.write(updated_student_data + "\n")

  file.close()

add_button.on_click(addFavoriteTVshow)
StudentNameSelection.observe(StudentInfoChange, names='value')
display(StudentNameSelection, email_text, courses_text, link_text,favorite_tvshow_text,add_button)
StudentInfoChange()


```
