```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
#Tozeret = "" # @param [""]
CarsNameSelection = widgets.Dropdown(
    options=data_df['tozeret_nm'].unique().tolist(),
    description='Tozeret:'
)
#Kinuy_Mishari = "" # @param [""]
CarsNickSelection = widgets.Dropdown(
    options=[],
    description='Kinuy Mishari:',
    style={'description_width': '115px'},  # Allocate more space for the label
    #layout=Layout(margin='10px 0 0 0')  # Add spacing between dropdowns
)

def update_dropdown_options(change=None):
    selected_name = CarsNameSelection.value
    car_nicknames = data_df[data_df['tozeret_nm'] == selected_name]['kinuy_mishari'].unique() # תראה לי את הדאטה כינוי רכב איפה שהשם של המכונית שווה לשם הנבחר בדרופ
    CarsNickSelection.options = car_nicknames.tolist()

output_area = Output() #שיטה של ווידג'ט להראות הדפסות

def show_details(change=None):
    selected_name = CarsNameSelection.value
    selected_nickname = CarsNickSelection.value
    selected_row = data_df[(data_df['tozeret_nm'] == selected_name) & (data_df['kinuy_mishari'] == selected_nickname)]

    total_records = selected_row.shape[0]
    unique_ramat_gimur = selected_row['ramat_gimur'].unique()

    unique_ramat_gimur_string = ', '.join(unique_ramat_gimur)

    # Update the output
    with output_area:
        output_area.clear_output()
        print(f"Total Cars: {total_records}")
        print(f"Unique Ramat Gimur: {unique_ramat_gimur_string}")

CarsNameSelection.observe(update_dropdown_options, names='value') # עושים אובסרב תמיד לפונקציה שהתלות קיימת בה (שרוצים לראות שינויים לפיה)
CarsNickSelection.observe(show_details, names='value')
display(CarsNameSelection)
display(CarsNickSelection)
show_details()
display(output_area)


import requests
import pandas as pd
pd.set_option("display.max_columns", None)
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)


# Create a DataFrame from the 'records' in the JSON response
data_df = pd.DataFrame(data['result']['records'])

# Set Tozeret with the 'tozeret_nm' column values
print("\n")
print(data_df['tozeret_nm'].tolist())
print(data_df['kinuy_mishari'].tolist())
```
