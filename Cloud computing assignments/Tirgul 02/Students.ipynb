#Imports

```python
from google.colab import drive
drive.mount('/content/drive')
import json

import ipywidgets as widgets
```

#Json file contents into python dictionary


```python
filepath = '/content/drive/MyDrive/Cloud computing - Daniel Ayash/Tirgul 02/Students.json'

with open(filepath, "r") as file:
  Students = json.load(file)
file.close()

for student in Students:
  if "favoriteTVshows" in student:
    print(student["name"],student["email"],student["courses"],student["link"],student["favoriteTVshows"])
  else:
    print(student["name"],student["email"],student["courses"],student["link"])
```

#Making of dropdown and content boxes (Json version)

```python
# @title Students form

StudentNameSelection = widgets.Dropdown(
    options = [student["name"] for student in Students],
    description='Student:',
    disabled=False,
)

email_text = widgets.Text(description='Email:', disabled=True)
courses_text = widgets.Text(description='Courses:', disabled=True)
link_text = widgets.Text(description='Link:', disabled=True)

favorite_tvshow_text = widgets.Text(description='TV Show:', disabled=False)
add_button = widgets.Button(description="Add TV Show")


def StudentInfoChange(change=None):
  student_name = StudentNameSelection.value
  for student in Students:
    if student["name"] == student_name:
      email_text.value = student["email"]
      fixed_courses = student["courses"].replace(" ", ",")
      courses_text.value = fixed_courses
      link_text.value = student["link"]

def addFavoriteTVshow(button):
  student_name = StudentNameSelection.value
  favorite_tv = favorite_tvshow_text.value

# iterates over every object in the dictionary (object = student)
  for student in Students:
    if student["name"] == student_name:
      if favorite_tv == "":
        if "favoriteTVshows" in student:
          del student["favoriteTVshows"]
      else:
         student.update({"favoriteTVshows":favorite_tv})


  with open(filepath, "w") as file:
    json.dump(Students,file,indent=4)

  file.close()

add_button.on_click(addFavoriteTVshow)
StudentNameSelection.observe(StudentInfoChange, names='value')
display(StudentNameSelection, email_text, courses_text, link_text,favorite_tvshow_text,add_button)
StudentInfoChange()


```
